# WeatherPy

```python
# import the dependencies
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from scipy.stats import linregress

# importing citipy to look for nearest city with geo coordinates
# https://pypi.org/project/citipy/
# also ran a pip install citypy at the command line
from citipy import citipy

# set the range of Lat and Long
lat_range = (-90, 90)
long_range = (-180, 180)

# importing the API key
from api_keys import weather_api

# set the measurment unit
temp_unit = "imperial"

# set the base URL for the API call
base_url = f"https://api.openweathermap.org/data/2.5/weather?"

# API query URL documented here:
# https://openweathermap.org/api/one-call-3
api_url = f"{base_url}&units={temp_unit}&appid={weather_api}&q="
#api_url - to test the url
```

```python
# creating the Cities list

# list for cities, latitudes/longitudes
lat_long = []
cities = []

# variables for random sets of latitude/longitude combinations
lats = np.random.uniform(lat_range[0], lat_range[1], size=1500)
longs = np.random.uniform(long_range[0], long_range[1], size=1500)
lat_long = zip(lats, longs)

# find the nearest city for reach latitude/longitude set
for geo_coords in lat_long:
    city = citipy.nearest_city(geo_coords[0], geo_coords[1]).city_name
    
    # add unique cities to our list
    if city not in cities:
        cities.append(city)
        
# check the # of cities with a count
len(cities)
```

```python
# check to ensure the cities are actually populated
print(f"{cities}")
```

```python
# Here's the section for the API calls

# setting the lists for each attribute
city_name = []
cloudiness = []
dates = []
humidity = []
lats = []
longs = []
max_temp = []
wind_speed = []
country = []

# declaring i for a count 
i = 0

# variables for iterations
record_num = 1
record_set = 1

# print start of data retrieval
print("\n-----------------------------------\nBeginning Data Retrieval\n-----------------------------------")

# for loop to run through each entry
for city in cities:
    
    # the API call
    response = requests.get(api_url + city).json()
    #print(response) - used this for debugging

    # message to let them know which record __ of set __ is being processed
    print(f'Processing Record {record_num} of Set {record_set} | {city}')
          # {response["name"]}')
    
    try:        
        # for matching cities, enter the data into the DF
        cloudiness.append(response['clouds']['all'])
        country.append(response['sys']['country'])
        dates.append(response['dt'])
        humidity.append(response['main']['humidity'])
        lats.append(response['coord']['lat'])
        longs.append(response['coord']['lon'])
        max_temp.append(response['main']['temp_max'])
        wind_speed.append(response['wind']['speed'])
        
        # make sets of 50 to help view the data
        if record_num > 49:
            record_set += 1
            record_num = 1
            
            city_name.append(response['name'])
        else:        
            # increase the record number count
            record_num += 1
            city_name.append(response['name'])
            
    except:
        print(f"City {city} not found. Skipping..." )
        record_num += 1
    
print("\n-----------------------------------\nData Retrieval Complete\n-----------------------------------")
```

# Convert Raw Data to DataFrame

```python
# create a dictionary to prepare for the dataframe
weather_dict = {
    "City": city_name,
    "Cloudiness":cloudiness,
    "Country":country,
    "Date":dates,
    "Humidity":humidity,
    "Lat":lats,
    "Long":longs,
    "Max Temp":max_temp,
    "Wind Speed":wind_speed
}
```

```python
# create the dataframe
weather_DF = pd.DataFrame(weather_dict)

# print first few rows as an example
weather_DF.head()
```

```python
# time to output our data

# path to save the file
output_file = "../Output/weatherpy_output.csv"

# export the data
weather_DF.to_csv(output_file, index_label="city_id")
```

## Linear Regression

```python
# create northern and southern hemisphere dataframes for linear regression

north_hem = clean_city_data.loc[clean_city_data["Lat"] >= 0]
south_hem = clean_city_data.loc[clean_city_data["Lat"] < 0]
```

### Northern Hemisphere - Temperature (F) vs. Latitude

```python
# Northern Hemisphere - Max Temp vs. Latitude Linear Regression: 
x_values = pd.to_numeric(north_hem['Lat']).astype(float)
y_values = pd.to_numeric(north_hem['Max Temp']).astype(float)
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
equation = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))

# plot 1 : Northern Hemisphere - Max Temp vs. Latitude Linear Regression
plt.scatter(x_values,y_values, edgecolors="navy", facecolors="violet")
plt.plot(x_values,regress_values,"r-")
plt.annotate(equation,(5,-10),fontsize=15,color="red")
plt.xlabel('Latitude')
plt.ylabel('Max Temperature (F)')
plt.title('Northern Hemisphere - Max Temp vs. Latitude Linear Regression\n2022-11-03')

# Calculated pvalue to determine whether correlation is significant(if there is truly a relationship):
print(f"The r-value is: {rvalue}")

plt.savefig("../Output/north_tempvLat.png", facecolor=fig_color.get_facecolor())
plt.show() 
```

### Southern Hemisphere - Temperature (F) vs. Latitude

```python
# Southern Hemisphere - Max Temp vs. Latitude Linear Regression: 
x_values = pd.to_numeric(south_hem['Lat']).astype(float)
y_values = pd.to_numeric(south_hem['Max Temp']).astype(float)
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
equation = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))

# plot 1 : Southern Hemisphere - Max Temp vs. Latitude Linear Regression
plt.scatter(x_values,y_values, edgecolors="navy", facecolors="violet")
plt.plot(x_values,regress_values,"r-")
plt.annotate(equation,(-55,77),fontsize=15,color="red")
plt.xlabel('Latitude')
plt.ylabel('Max Temperature (F)')
plt.title('Southern Hemisphere - Max Temp vs. Latitude Linear Regression\n2022-11-03')

# Calculated pvalue to determine whether correlation is significant(if there is truly a relationship):
print(f"The r-value is: {rvalue}")

plt.savefig("../Output/south_tempvLat.png", facecolor=fig_color.get_facecolor())
plt.show() 
```

#### Analysis - Latitude vs. Max Temp Linear Regression for Northern & Southern Hemispheres

* The Max Temps rise the closer you get to the Equator (Latitude 0) for both Northern and Southern Hemispheres.
* The slopes move in opposite directions when you compare the Northern and Southern Hemispheres and their relative Latitudes
* The Southern Hemisphere has more of a scattered data set for Max Temps than the Northern Hemisphere.
